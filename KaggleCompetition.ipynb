### numerical and categorical columns

```python
target_feature = "ACCLASS"
predictors = [column for column in df.columns if column not in [target_feature]]
numerical_columns = ["YEAR", "MONTH", "DAY", "HOUR", "LATITUDE", "LONGITUDE"]
binary_category_columns = [column for column in predictors if df[column].nunique()==1]
multi_category_columns = [column for column in predictors if column not in numerical_columns + binary_category_columns]
```

```python
# now lets discard columns "DATE", "TIME" in multi_category_columns as their data are already extracted to the HOUR, DAY, MONTH
# FATAL_NO can also be discarded as it give no insight
# Neighbour hood is also discarded because it data is already given by HOOD_158 and HOOD_140
for column_name in ["DATE", "TIME", "FATAL_NO",'NEIGHBOURHOOD_158','NEIGHBOURHOOD_140']:
    multi_category_columns.remove(column_name)
```

# Handling Missing Values

```python
# checking missing values for numerical columns
df[numerical_columns].isnull().sum()
```

```python
# checking missing values for binary_category_columns
df[binary_category_columns].isnull().sum()
```

```python
# imputing null rows with "No"
df[binary_category_columns] = df[binary_category_columns].fillna("No")
```

```python
df[binary_category_columns].isnull().sum()
```

```python
df[multi_category_columns].isnull().sum()
```

```python
df[multi_category_columns] = df[multi_category_columns].fillna("Unknown")
```

```python
df[multi_category_columns].isnull().sum()
```

```python
# Ordinal columns
ordinal_columns = ['INVAGE', 'INJURY', 'HOOD_158', 'HOOD_140','DIVISION','VEHTYPE']
for column in ordinal_columns:
    multi_category_columns.remove(column)
print(ordinal_columns)
print(multi_category_columns)
```

```python
x = df[numerical_columns + binary_category_columns + multi_category_columns + ordinal_columns]
y = df[target_feature]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10086, stratify=y) #stratified sampling based on the target
print(f"x_train.shape: {x_train.shape}")
print(f"x_test.shape: {x_test.shape}")
print(f"y_train.shape: {y_train.shape}")
print(f"y_test.shape: {y_test.shape}")
```

### Encoding and Scaling data

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder ,StandardScaler, LabelEncoder, OrdinalEncoder

# onehot encoding for nominal columns
pipeline_categorical = Pipeline([
  ('onehot', OneHotEncoder(drop = "first", handle_unknown = "ignore")),
])

#standard scaling for numerical columns
pipeline_numerical = Pipeline([
  ('scaler', StandardScaler()),
])

#ordinal encoding for ordinal columns
pipeline_ordinal = Pipeline([
  ('ordinalencoder', OrdinalEncoder(handle_unknown = "use_encoded_value", unknown_value = -1)),
])

# full pipeline of all the encodings and scaling
pipeline_full = ColumnTransformer([
  ("categorical", pipeline_categorical, binary_category_columns + multi_category_columns),
  ("numerical", pipeline_numerical, numerical_columns),
  ("ordinal", pipeline_ordinal, ordinal_columns ),
])


pipeline_full.fit(x_train)

# transformed trainset
x_train_CT = pipeline_full.transform(x_train)

# transformed testset
x_test_CT = pipeline_full.transform(x_test)

print(f"x_train shape: {x_train.shape}")
print(f"x_train transformed shape: {x_train_CT.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"x_test transformed shape: {x_test_CT.shape}")
```

###  KNN Classifier with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

n_neighbors =  [3, 5, 7, 9]  # Number of neighbors to consider
weights = ['uniform', 'distance'] # Weighting of neighbors ('uniform' or 'distance')
p = [1, 2] # Power parameter for the Minkowski metric (1 for Manhattan distance, 2 for Euclidean distance)

parameter_KNeighbors = {'n_neighbors' : n_neighbors, 'weights' : weights, 'p' : p}
knnG = KNeighborsClassifier(algorithm = 'auto')

grid_search_KNeighbors = GridSearchCV(knnG, parameter_KNeighbors, cv=4, scoring='roc_auc', refit = True, n_jobs = -1, verbose = 2)
grid_search_KNeighbors.fit(x_train_CT,y_train)

KNeighbors_best_params_ = grid_search_KNeighbors.best_params_
KNeighbors_best_estimators_ = grid_search_KNeighbors.best_estimator_
print(KNeighbors_best_params_)
print(KNeighbors_best_estimators_)
print(KNeighbors_best_estimators_.score(x_train_CT, y_train))
```

```python
# check accuracy on test dataset
from sklearn.metrics import accuracy_score
KNeighbors_Estimators = KNeighbors_best_estimators_
y_predicted = KNeighbors_Estimators.predict(x_test_CT)
print(accuracy_score(y_test, y_predicted))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
# precision, recall and f1 score
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_predicted, pos_label='Non-Fatal Injury')
recall = recall_score(y_test, y_predicted, pos_label='Non-Fatal Injury')
f1 = f1_score(y_test, y_predicted, pos_label='Non-Fatal Injury')
print(precision)
print(recall)
print(f1)
```

```python
# Function to Plot the AUC Curve
def plot_Parameters_Vs_Scores(criterion_estimators,roc_auc_values,plotType):
    if plotType == "Scatter":
        plt.scatter(criterion_estimators, roc_auc_values)
    if plotType == "Line":
        plt.plot(criterion_estimators, roc_auc_values) 
    plt.xlabel("Parameters Vs AUC Scores")
    plt.legend(loc="best")
    plt.grid(True)
```

```python
# AUC for KNN
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score

criterion_estimators_2 = [1, 2, 3, 4, 5]

roc_auc_values = []

for item in criterion_estimators_2:
    knnG_clf = KNeighborsClassifier(algorithm = 'auto')
    knnG_clf.fit(x_train_CT, y_train)

    y_probas_trees = cross_val_predict(knnG_clf,x_train_CT, y_train, cv=4, method="predict_proba")
    y_tree_scores = y_probas_trees[:, 1] # score = proba of positive class
    roc_auc_trees = roc_auc_score(y_train,y_tree_scores)
    roc_auc_values.append(roc_auc_trees)
    
plot_Parameters_Vs_Scores(criterion_estimators_2, roc_auc_values,"Line")
plt.show()
```
